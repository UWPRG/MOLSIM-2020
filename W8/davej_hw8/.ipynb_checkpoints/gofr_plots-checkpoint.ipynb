# LETS PLOT SOME DATA

```python
import matplotlib
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

#### First lets plot the water radial distribution function 

```python
water = pd.read_csv('gor_water.dat', header=None, sep=' ')
```

```python
water = water.rename({0:'r', 1:'g(r)', 2:'unknown'}, axis=1)
```

```python
plt.figure(figsize=(5,2),dpi=200)
plt.plot(water['r'], water['g(r)'])
plt.xlim(0,6)
plt.title('Water radial distribution function', fontsize=8)
plt.xlabel('r (angstrom)', fontsize=8)
plt.ylabel('g(r)', fontsize=8)
```

### I calculated a couple other rdf's to try and understand what's going on. The peak at 1A definitely confused me.

oxygen-oxygen rdf

```python
oxy = pd.read_csv('gor_oxy.dat', header=None, sep=' ')
oxy = oxy.rename({0:'r', 1:'g(r)', 2:'unknown'}, axis=1)

plt.figure(figsize=(5,2),dpi=200)
plt.plot(oxy['r'], oxy['g(r)'])
plt.xlim(0,6)
plt.title('Oxygen radial distribution function', fontsize=8)
plt.xlabel('r (angstrom)', fontsize=8)
plt.ylabel('g(r)', fontsize=8)
```

Interesting, we cant really see much of a solvation shell when just looking at oxygens. But the distances make sense. 

This tells me the first peak in the water-water g(r) definitely corresponds to water-hydrogen covalent bond length. Super strong peak because every oxygen is bonded to 2 hydrogens forever.
