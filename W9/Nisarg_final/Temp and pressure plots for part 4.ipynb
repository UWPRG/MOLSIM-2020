Plotting the temperature for both the NVT files

```python
import numpy as np
import matplotlib.pyplot as plt

time, temp1 = [], []

with open("thermo_prod_nvt_273.xvg") as f:
    for line in f:
        cols = line.split()

        if len(cols) == 2:
            time.append(float(cols[0]))
            temp1.append(float(cols[1]))
```

```python
temp2 = []
with open("thermo_prod_nvt_500.xvg") as f:
    for line in f:
        cols = line.split()

        if len(cols) == 2:
            temp2.append(float(cols[1]))
```

```python
plt.figure()
plt.plot(time, temp1, c='r', label='temp = 273K')
plt.plot(time, temp2, c='b', label='temp = 500K')
plt.xlabel('Time (ps)')
plt.ylabel('Temperature (K)')
plt.title('Temperature plot for NVT')
plt.legend()
```
